_If your student number is even (i.e. it ends with 0, 2, 4, 6, or 8) then you start at (1,5), else you start at (5,3). ## Question IV: control flow

_Draw what is printed on the screen after the execution of the following code_

```python
def drawShape(size):
  s = ""
  for i in range (0,size):
    for j in range(0,size):
      if j < i or j > size - i - 1:
        s = s + " "
      else:
        s = s + "#"
    s += "\n"
  print(s)

drawShape(9)
```
